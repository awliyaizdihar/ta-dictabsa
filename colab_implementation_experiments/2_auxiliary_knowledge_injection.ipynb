# Import Library

```python
# Import Library Python
import pandas as pd
```

# Load Dataset

## Dataset Ulasan

```python
file_train = f'https://docs.google.com/spreadsheets/d/17CTFM7JGIFJVVaKzMx3ffm9OfbDXqGY5/gviz/tq?tqx=out:csv'
file_test = f'https://docs.google.com/spreadsheets/d/15vGpggHH_uyprLH2cM1RCTfRGLttzKuu/gviz/tq?tqx=out:csv'
file_validation = f'https://docs.google.com/spreadsheets/d/1QjmHbXVURLoYQafheq6buuy4vC9egGnG/gviz/tq?tqx=out:csv'
train_set = pd.read_csv(file_train)
test_set = pd.read_csv(file_test)
validation_set = pd.read_csv(file_validation)
```

## Dataset All Knowledge

```python
file_know = f'https://docs.google.com/spreadsheets/d/1ysks_5UHePOpViY2qghotY-xqig3BO-pEIkFtkrbCR0/gviz/tq?tqx=out:csv'
know_set = pd.read_csv(file_know, usecols=['aspect', 'knowledge'])
know_set
```

# Mengekstrak Data

Dataset masih berupa kumpulan ulasan dari peneliti sebelumnya.

Data tersebut akan diekstrak menjadi dataset dengan 3 kolom (review, aspect, dan sentiment).

```python
test_str = train_set.review[1]
test_str
```

## Mengekstrak Aspek

Aspek dalam ulasan diapit di antara tanda [B-ASP] dan [E-ASP].

```python
datasets = [train_set, validation_set, test_set]
```

```python
# menginisialisasi substring
b_asp = "[B-ASP]"
e_asp = "[E-ASP]"

# mengambil kata aspek di antara 2 substring menggunakan split() dan join()
for sets in datasets:
  sets['aspect'] = sets.review.apply(lambda x: ''.join(x.split(b_asp)[1].split(e_asp)[0]))
```

## Mengekstrak Sentimen

Polaritas sentimen terletak di akhir kalimat ulasan.

```python
for sets in datasets:
  new = sets['review'].str.split(pat=" $LABEL$ ", n=1, expand=True, regex=False)
  sets['review'] = new[0]
  sets['sentiment'] = new[1]
```

# Praproses Data

## Encoding Label Sentimen

Mengubah label sentimen yang berupa kata (Positive, Neutral, Negative) menjadi angka (1, 0, -1).

```python
# mengubah label sentimen manjadi angka
for sets in datasets:
  sets.replace({'sentiment' : {'Positive': 1 , 'Neutral': 0, 'Negative':-1}}, inplace=True)
```

## Normalisasi Kata Aspek

Mengembalikan aspek ke kata dasar. (Contoh: "pemandangannya" ==> "pemandangan")

```python
# menghapus imbuhan -nya pada kata aspek
for sets in datasets:
  sets['aspect'] = sets['aspect'].str.replace("nya$", "", regex=True)
```

# Inserting Knowledge

```python
def insert_know(dataset):
  dataset['before_knowledge'] = dataset.review.apply(lambda x: x[:x.find(b_asp)])
  dataset['after_knowledge'] = dataset.review.apply(lambda x: x[x.find(e_asp) + len(e_asp):])
  dataset['knowledge'] = ' $ ' + dataset['knowledge'] + ' $'
  dataset['review'] = dataset[['before_knowledge', 'aspect', 'knowledge', 'after_knowledge']].stack().groupby(level=0).agg(''.join)
  dataset.drop(columns=['before_knowledge', 'knowledge', 'after_knowledge'], inplace=True)
```

```python
train_know = []
train_know = pd.merge(train_set, know_set, on='aspect', how='left')
insert_know(train_know)
train_know
```

```python
validation_know = []
validation_know = pd.merge(validation_set, know_set, on='aspect', how='left')
insert_know(validation_know)
validation_know
```

```python
test_know = []
test_know = pd.merge(test_set, know_set, on='aspect', how='left')
insert_know(test_know)
test_know
```

# Concatenating Knowledge

Menghapus penanda posisi aspek yang sudah tidak terpakai

```python
test_set
```

```python
# menghapus tanda aspek yang sudah tidak terpakai
for sets in datasets:
  sets['review'] = sets['review'].str.replace(b_asp, "", regex=False)
  sets['review'] = sets['review'].str.replace(e_asp, "", regex=False)
```

Concatenating Knowledge

```python
train_concat = []
train_concat = pd.merge(train_set, know_set, on='aspect', how='left')
train_concat['review'] = train_concat[['review', 'knowledge']].stack().groupby(level=0).agg(' [SEP] '.join)
train_concat.drop(columns=['knowledge'], inplace=True)
train_concat
```

```python
validation_concat = []
validation_concat = pd.merge(validation_set, know_set, on='aspect', how='left')
validation_concat['review'] = validation_concat[['review', 'knowledge']].stack().groupby(level=0).agg(' [SEP] '.join)
validation_concat.drop(columns=['knowledge'], inplace=True)
validation_concat
```

```python
test_concat = []
test_concat = pd.merge(test_set, know_set, on='aspect', how='left')
test_concat['review'] = test_concat[['review', 'knowledge']].stack().groupby(level=0).agg(' [SEP] '.join)
test_concat.drop(columns=['knowledge'], inplace=True)
test_concat
```
